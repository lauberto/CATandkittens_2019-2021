Посмотрим, насколько хорошо у нас получается определять склонение

```python
def get_declension(noun_lemma, gender):
    last_letter = noun_lemma[-1]
    if last_letter == 'а':
        return '1а'
    elif last_letter  == 'я':
        return '1я'
    elif last_letter in 'оейцкнгшщзхфвпрлджчсмтб':
        return '2'
    elif last_letter == 'ь':
        if noun_lemma[-2] in 'чжшщ' or gender == 'Fem':
            return '3'
        else:
            return '2'
```
