```python
with open('preposions_CAT.txt', encoding='utf-8') as f:
    verbs_p = f.read().split()
```

```python
verbs_p_counter = Counter(verbs_p)
```

```python
verbs = {}
for i in verbs_p_counter:
    [verb, preposition] = i.split('_')
    if verb not in verbs:
        verbs[verb] = {}
    if preposition not in verbs[verb]:
        verbs[verb][preposition] = verbs_p_counter[i]
    else:
        verbs[verb][preposition] += verbs_p_counter[i]       
```

```python
def get_preposition_data(verb):
    if verb in verbs:
        return verbs[verb]
    else:
        return {}
```
